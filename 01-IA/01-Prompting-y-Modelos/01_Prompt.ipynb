# De que hablamos cuando hablamos de Ingenieria de Prompt

La idea principal de esta disciplina es poder dar instrucciones de forma presisa a un modelo de LLM para obtener la respuesta esperada

# 🧠 ¿Qué es un Prompt?

Un prompt es la instrucción que le das a un modelo de IA para que genere una respuesta.

En herramientas como ChatGPT, Gemini, etc., el prompt puede ser:

- una pregunta  
- una orden  
- un contexto + tarea  
- un ejemplo  

---

## ⚠️ Importancia del prompt

Es fundamental comprender cómo redactar un prompt porque los modelos LLM (Large Language Models) no “piensan” como humanos:

- No entienden intención implícita  
- No adivinan contexto  
- No corrigen ambigüedad automáticamente  

> Responden exactamente a lo que les pedís (y cómo se los pedís)

---

# ❓ ¿Todos los prompts son iguales?

Un prompt es una instrucción que le damos al modelo para obtener una salida.

Sin embargo, los prompts varían según la función del sistema donde se utilice el LLM:

- Chatbots conversacionales  
- Agentes autónomos  
- Sistemas de control o automatización  
- Generadores de imágenes  

En general, los prompts mantienen una estructura similar, pero se adaptan a la tarea específica del sistema.

---

# 🧩 Estructura de un Prompt Efectivo

Un prompt efectivo no es solo una pregunta o instrucción aislada.

Es una composición estructurada de elementos que guían al modelo para generar una respuesta precisa, útil y controlada.

---

## ⚙️ Fórmula general

[Rol] + [Tarea] + [Contexto] + [Formato de salida] + [Restricciones]



---

# 🎭 1. Rol (opcional)

Define “quién” debe ser el modelo al responder.

### Ejemplos

- Actuá como un científico de datos  
- Respondé como un asesor financiero  
- Sos un profesor universitario  

### Objetivo

- Ajustar tono  
- Definir nivel técnico  
- Cambiar enfoque de la respuesta  

---

# 📌 2. Tarea (obligatorio)

Es la instrucción principal: qué debe hacer el modelo.

### Ejemplos

- Explicá qué es el overfitting  
- Contame una historia de hadas  
- Resumí este texto  

### Objetivo

Debe ser clara, directa y sin ambigüedad.

---

# 📚 3. Contexto

Aporta información adicional para guiar la respuesta.

### Ejemplos

- En el contexto de modelos agrícolas en Argentina  
- Según el reglamento del colegio  

### Objetivo

- Reducir ambigüedad  
- Mejorar precisión  
- Adaptar la respuesta al dominio  

---

# 📋 4. Formato de salida

Define cómo querés que se estructure la respuesta.

### Ejemplo

- Definición
- Causas
- Ejemplo

### Objetivo

En automatización o agentes es crítico para estandarizar la salida.

---

# 🚫 5. Restricciones

Limitan o condicionan la respuesta.

### Ejemplos

- No uses lenguaje técnico  
- Máximo 3 líneas  
- No inventes información  

### Objetivo

Controlar calidad, longitud y comportamiento del modelo.

---

# 🧪 Ejemplo de Prompt Completo

Actuá como un científico de datos.

Explicá qué es el overfitting en machine learning,
en el contexto de predicción agrícola en Argentina.

Respondé en formato:

definición

# 💡 Idea clave

Un buen prompt no es una pregunta mejor escrita,  
es una instrucción estructurada diseñada para guiar la salida del modelo.

No existe el prompt perfecto a la primera.

El proceso real es:

- Diseñar prompt
- Evaluar respuesta
- Ajustar
- Volver a probar


Podés evaluar un prompt según:

- precisión
- claridad
- consistencia
-cumplimiento de formato




## TECNICAS DE PROMPTING

- **Role prompting**

Consiste en asignarle un rol explícito al modelo para condicionar su comportamiento.

>Prioriza cierto tipo de conocimiento

>Ajusta nivel técnico

>Cambia el enfoque de resolución

Usar esta tecnica cuando:

Buscas precisión técnica

Cambio de perspectiva

Mismo problema distintos enfoques


El modelo no “se convierte” en ese rol,
solo ajusta probabilísticamente su respuesta.

El rol reduce ambigüedad sobre cómo responder, no sobre qué responder.


---------


- **Few-shot prompting**

Es la tecnica de poner ejemplos dentro del prompt para poder darle un formato al modelo de como queremos que nos responda

-Mejora consistencia

-Define formato implícito

-Reduce errores en tareas repetitivas


Dentro del mismo prompt agregamos

Input: trigo → Output: cultivo de invierno

Input: "El documento entregado es una forma F011.2"  nombre_archivo:"F011.2"


Se utiliza esta tecnica principalmente para

Clasificación

Transformación de datos

Formatos específicos

Tareas semi-estructuradas

---------------
- **Chain-of-thought**

Chain-of-Thought prompting consiste en inducir al modelo a generar pasos intermedios de razonamiento antes de llegar a una respuesta final.

Forzar al modelo a razonar paso a paso.

Mejora resolución de problemas complejos

Reduce errores lógicos

Hace explícito el razonamiento


Es una herramienta muy util ya que modelos como ChatGPT no razonan como humanos.

Funcionan prediciendo la siguiente palabra/token.

Al pedir un “paso a paso”, lo que hacés es:

Aumentar el espacio de generación

Obligar a descomponer el problema

Reduce saltos probabilísticos incorrectos


Podemos optar por una version guiada en la cual estructuramos paso a paso de forma explicita como debe de realizar el procedimiento, como ejemplo:

"
1. Identificá variables

2. Aplicá fórmula

3. Calculá resultado

"

Otra tecnica es la inducción del proceso de pensamiento por pasos

No citas los pasos, pero los inducís:

>“Pensá cuidadosamente antes de responder y devolvé solo el resultado final”



Es importante entender que esta tecnica no garantiza mejor conocimiento ni que los pasos intermedios que realice sean correctos, pero al segmentar los pasos es mas simple poder evaluar la respuesta de cada paso y encontrar sus fallas

---------------
- **Structured prompting**

Definir explícitamente el formato de salida esperado.

- Hace la salida predecible
- Facilita automatización
- Reduce ambigüedad
