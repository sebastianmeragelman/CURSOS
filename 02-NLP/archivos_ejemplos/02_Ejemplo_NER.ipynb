Hay varios paquetes disponibles para descargar de spacy, mayor sea el tamaño suele funcionar mejor, pero tambien implica un mayor costo de recursos para su utilizacion

es: El idioma (Español).

core: El tipo de capacidades (en este caso, un modelo central que incluye vocabulario, sintaxis, entidades, etc.).

news: El tipo de texto con el que fue entrenado (noticias, artículos, web).

sm: El tamaño del modelo (small, mdedium, lgarge, trf Transformer).

```python
# Importamos la libreria
import spacy

# Cargar modelo en español
nlp = spacy.load("es_core_news_sm")

# Texto de ejemplo
texto = "Juan Perez pagó la factura 1234 por $1500 en Córdoba el 12/05/2024"

# Procesar texto
doc = nlp(texto)

print("🔎 Entidades detectadas:\n")

for ent in doc.ents:
    print(f"Texto: {ent.text} | Tipo: {ent.label_}")
```

```python

from spacy.training.example import Example

nlp = spacy.blank("es")
ner = nlp.add_pipe("ner")

ner.add_label("FACTURA")
ner.add_label("CLIENTE")
ner.add_label("MONTO")

optimizer = nlp.begin_training()

for i in range(20):
    for text, annotations in train_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], sgd=optimizer)
```

```python

# Vamos a realizar sobre un texto de ejemplo el NER
doc = nlp("Factura 9999 cliente ACME monto $3000")

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
texto = "Juan Perez works at Google and lives in Argentina"

# 1. Tokenización
tokens = nltk.word_tokenize(texto)

# 2. POS tagging (etiquetado gramatical)
pos_tags = nltk.pos_tag(tokens)

# 3. NER
arbol = nltk.ne_chunk(pos_tags)

print(arbol)
```
